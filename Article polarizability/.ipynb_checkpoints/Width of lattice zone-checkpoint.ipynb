```python
# constants
k_b = 1.3806488e-23
h_bar = 1.054571726e-34
m_at = 1.66053892e-27
k = 2*pi/(0.5*1e-6)
m = 169
c = 299792458 # m\s
ϵ0 = 8.854187817e-12 #F/m
α0 = 1.6487772731e-41 #C2 m2 J-1
λ = 0.532e-6
w0 = 60e-6 # m, waist
h = 2*pi*h_bar
λlatt = 0.8e-6
depth = 10
```

Recoil energy and freqyency

```python
Er = h_bar**2 * (2*pi/λlatt)**2/(2*m*m_at)
f_rec = Er/h
```

For different lattice depth calculate band structure

```python
n_points = 50
all_data = {}
for depth in  [2,5,10,20,30,40,50,70,100,150]:
    qs = linspace(-1,1,101)
    eagenvalues = []
    for q in qs:
        ks = 2*arange(-n_points,n_points+1) + q
        H = diag((ks)**2 + depth/2,0) - diag([depth/4]*(2*n_points),1)- diag([depth/4]*(2*n_points),-1)
        eigval, eigvec = linalg.eigh(H)
        eagenvalues.append(eigval)
    data = list(zip(*eagenvalues))
    all_data[depth]=data
```

```python
# construct list fw of band widths
widthes = {}
for key in sorted(all_data.keys()):
    widthes[key] = [max(x)-min(x) for x in all_data[key]]
#     print(key,'\t',[max(x)-min(x) for x in all_data[key][:5]])
fw = list(zip(*[widthes[key] for key in sorted(widthes.keys())]))
```

```python
for n in range(4):
    plot(sorted(widthes.keys()),array(fw[n])*Er/h,'-*',label=n)
yscale('log')
ylabel('band widht, Hz')
xlabel('lattice depth, Er')
legend()
```
