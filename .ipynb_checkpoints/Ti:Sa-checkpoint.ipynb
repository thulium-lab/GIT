### Расчет финеса и усиления интенсивности

```python
def powerRes(R1,R2):
    """Increace of power inside the cavity"""
    return (1-R1)*(1+sqrt(R2))**2 / (1 - sqrt(R1*R2))**2
def powerRes2(R1,R2):
    """Increace of power inside the cavity"""
    return (1-sqrt(R1))**2*(1+sqrt(R2))**2 / (1 - sqrt(R1*R2))**2
def finesse(R1,R2):
    "Cavity finesse"
    return pi * (R1*R2)**0.25 / (1 - sqrt(R1*R2))
```

### Расчет частоты колебаний в красной ловушке

```python
a0 = ct.value('atomic unit of length')
m0 = ct.value('atomic mass constant')
m_Tm = 169
wl_mag = 814e-9
w0 = 80e-6
a = 170# a.u.
p=1# watt
```

```python
def f_r_dip(a, w0,p,wl=800e-9):
    m_Tm = 169
    return 2/pi/w0**2 * sqrt(a0**3*a*p/(ct.c*m0*m_Tm))
def f_z_dip(a,w0,p,wl):
    return wl / (sqrt(2)*pi*w0)*f_r_dip(a, w0,p)
def f_r_lattice(a,w0,p,wl=800e-9):
    return 2*f_r_dip(a, w0,p,wl)
def f_z_lattice(a,w0,p,wl):
    return sqrt(2)*pi*w0/wl * f_r_lattice(a,w0,p,wl)
def lattice_depth_uK(a,w0,p,wl):
    return 2*pi*a0**3 /ct.c * a * 2*4*p/(pi*w0**2) / ct.k *1e6
```

```python
print('f_r_lattice ',f_r_lattice(a,w0,p,wl_mag))
print('f_z_lattice ',f_z_lattice(a,w0,p,wl_mag))
```

```python
nu_recoil = ct.hbar * (2*pi/wl_mag)**2/ (m0*m_Tm)
print('nu recoil=%.0f Hz' %nu_recoil)
```

```python
e_recoil = ct.hbar**2*(2*pi/wl_mag)**2/(2*m0*m_Tm)
e_recoil_uK = e_recoil / ct.k *1e6
print('Er = %.0f Hz' % (e_recoil / ct.h))
print('Er = %.0f nK' % (e_recoil / ct.k *1e9))
```

### Functions - it's better to move them to library

```python
def m_l(l):
    """Free space matrix, l - optical path [meters] = d*n, n-reflactive index, d-real distance"""
    return array([[1,l],
                  [0,1]])

def m_f(f):
    """Thin lens matrix, f - rear focal distance [meters]"""
    return array([[1,0],
                  [-1/f,1]])

def z_r(w0,lda):
    """Calculates rayleigh range [m], w0 - waist [m], lda - wavelength [m]"""
    return (pi*w0**2/lda)

def q(z,zr=None,w0=None,z0=0):
    """calculates q-parameter,z - coordinate [m] (distance to the waist position if z0=0), z_r - rayleigh range [m]
    or waist radios w0 [m], z0 - distance to the waist"""
    if zr == None:
        zr = z_r(w0,lda)
    return (z-z0)+1j*zr

def w_z(z,lda,zr=None,w0=None,z0=0):
    """calculates beam radius [m] based on z_r - rayleigh range [m] or waist radios w0 [m],lda - wavelength [m]"""
    if zr == None:
        zr = z_r(w0,lda)
    return sqrt(lda / pi * zr)*sqrt(1 + (z-z0)**2/zr**2)

def get_w(q,lda):
    """Returns beam radius w [m] based on q-parameter"""
    return sqrt(lda / (pi * imag(-1/q)))

def get_w0(q,lda):
    """Returns beam waist radius w [m] based on q-parameter"""
    return sqrt(lda / pi * imag(q))

def waist_and_pos(z,z0,w0):
    return waist(z-z0,w0)

def finesse (R,A):
    """R - input mirror reflectivity, A - losses in the cavity"""
    return pi/(1 - sqrt(R*(1-A)))

def apply_m_to_q(m,q):
    """Apply matrix m to q-parameter"""
    return (m[0,0]*q + m[0,1]) / (m[1,0]*q + m[1,1])

def matrix_mult(*mts):
    """Multiplies matrixes mts: m1*m2...*mN"""
    if len(mts) > 1:
        res = mts[0]
        for m in mts[1:]:
            res = dot(m,res)
        return res
```

### Main class, that construct beam trace based on matrexes. Also better to move to library

```python
class BeamTrace:
    def __init__(self,matrexes,q_in,z0=0,n_points=10):
        self.n_points = n_points
        self.z0 = z0 #distance from q_in point to 0, needed only for convinience
        self.matrexes = matrexes # array of matrixes or labels - strings (names) at wich to calculate q-parameter
        self.xs = [] # x coordinates
        self.ws = [] # beam waists vs. xs
        self.qz_to_print = [] # future array of (label,q) for labels in matrexes
        self.q_in = q_in # initial q-parameter of the beam
        
    def constructRey(self):
        """Function that construct waists vs x posision"""
        self.xs = []
        self.ws = []
        self.qs_to_print = []
        q_in = self.q_in
        direction = 1 # direction of the beam (if mirror comes it changes the direction)
        for M in self.matrexes:
            
            if type(M) == str: # it is a label to save current beam parameter
                    self.qs_to_print.append((M,get_w(q_in,lda),q_in))
                    print(self.qs_to_print[-1])
                    continue
                    
            if M[0][1] != 0: # it is free space matrix 
                xs = linspace(0,M[0][1],self.n_points)
                ws = w_z(xs+real(q_in),lda=lda,zr=imag(q_in)) # calculates waists
                #print(ws)
                
                if self.xs != []: # point where if stoped in previous matrix
                    extr = self.xs[-1]
                else :
                    extr = self.z0
                    
                self.xs.extend(direction*xs+extr)
                self.ws.extend(ws)
                
            elif M[0][1] == 0 and M[1][0] == 0: # ones matrix means a mirror - change of the direction
                direction *= -1
                continue
            q_in = apply_m_to_q(M,q_in) #calculate new q-parameter for the next matrix
        
```

### Для усиливающего резонатора (код ниже)

1. Размер перетяжки почти не зависит от кривизны зеркал, но очень сильно зависит от расстояния от центра до зеркала: при r-1 мм это 63 мкм, при r-1см это 110 мкм.
2. Для 100 мкм перетяжки (зеркала в 244 мм от центра) оптимально ставить согласующую линзу 50 см на расстоянии 60 см от центра
3. Толщина стекол камеры 6.4 мм, а т.к. в матричной оптике входит путь d/n, то эффективное расстояние до центра меньше на 2 мм чем физическое, т.е. надо ставить зеркала на 2 мм дальше, чем в расчетах
4. Брюстеровская пластинка не только смещает пучок, но и вносит астигматизм. Смещение мучка можно найти по формуле
$\Delta = \frac{d sin(i-i')}{cos(i')}$, что равно 0.4 мм для пластинки толщиной 1 мм. Астигматизм приведет к смещению перетяжки примерно на 0.3 мм и изменению ее размера на 1.5%.

```python
# Находим размер перетяжки (выводим в мкм) в зависимости от d.

r=.25# mirror curvature
d = r-0.01 # distance from center to mirrors

cavity = matrix_mult(m_l(d),m_f(r/2),m_l(2*d),m_f(r/2),m_l(d)) # cavity matrix
res = root(lambda x: abs(apply_m_to_q(cavity,1j*x)-1j*x),0.05)
w0 = w_z(z=0,lda=lda,zr=res.x[0])
print(w0*1e6)
```

```python
ht = 0.25 # m, from cloud to upper plain mirror
hl = 0.3 # m, from cloud to tupper lens
hb = 0.2 # m, from cloud to bottom curved mirror
fl = 0.2 # m, lens focal distance
fb = 0.2 # m, corresponds to x2 curvature
```

```python
def unit(fl=fl,fb=fb,hl=hl-0.003,hb=hb):
    r = matrix_mult(m_l(ht-hl),m_f(fl),m_l(hl+hb),m_f(fb),m_f(fb),m_l(hl+hb),m_f(fl),m_l(ht-hl))
    zz1 = linspace(0,ht-hl,100)
    zr = abs(sqrt((r[0,0]-r[1,1])**2 + 4*r[0,1]*r[1,0] + 0j)/ (2*r[1,0]))
    ws1 = w_z(zz1,zr,λ)
    zz2 = linspace(ht-hl,ht+hb,100)
    q1 = apply_m_to_q(m_l(ht-hl),1j*zr)
    q2 = apply_m_to_q(m_f(fl),q1)
    ws2 = w_z(zz2-zz2[0]+real(q2),imag(q2),λ)
    zz3 = linspace(ht+hb,ht+hb+2*hb,100)
    q3 = apply_m_to_q(m_l(hb+hl),q2)
    q4 = apply_m_to_q(m_f(fb),q3)
    ws3 = w_z(zz3-zz3[0]+real(q4),imag(q4),λ)
    print("Waist at cloud position %.1f um" % (w_z(ht-zz2[0]+real(q2),imag(q2),λ)*1e6))
    plot(zz1,ws1)
    plot(zz2,ws2)
    plot(zz3,ws3)
    ylim(bottom=0)
    vlines(ht,0,1)
#     print(apply_m_to_q(m_l(hl),q2))
interact(unit,fl=(0,1,0.05),fb=(0,1,0.05),hl=(0.15,ht,0.001),hb=(0.15,0.3,0.001))
```
