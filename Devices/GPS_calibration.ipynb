```python
# name of the file to read data from
d_f = r'D:\\!Data\\gps_data\\data_log_10MHz_24_10_2017.txt'
```

```python
all_data=[]
last = 1e5
with open(d_f,'r') as f:
    for line in f:
        s = line.split(',')
        try:
            res = (int(s[9]),int(s[7]),int(s[8]),float(s[12].split('=')[1])) # (n,t_cpu,dt,df)
        except IndexError: # sometimes some fields a unavailable, skip it
            res = (int(s[9]),int(s[7]),int(s[8]),0) # (n,t_cpu,dt,df)
        if res[0] < last: # new measurement seria started
            all_data.append({'date':datetime.strptime(s[0].split(' ')[0],'%Y-%m-%d'), # date of meas.
                             't_start':s[0], # time of first folder (to distinguish)
                             'freq':int(s[3].strip().split(' ')[0]), # measurement frequency in kHz
                             'data':[res]}) # data as specified above
        else:
            all_data[-1]['data'].append(res)
        last = res[0]
print('len', len(all_data))
```

```python
for dd in all_data:
#     print(dd['data'])
    d = array(dd['data'],dtype=[('n','>i4'),('t_cpu','>i4'),('dt','>i4'),('df','>f4')])
    print('df=%.1e err=%.1e for ' % (d['df'][-1],1 / (dd['freq']*1e3 * d['t_cpu'][-1])), str(dd['freq']/1e3) + '  ' + dd['t_start'])
    plot(d['n'],d['df'])
xlabel('N shots (x100s)')
ylabel('Rel. freq. deviation')
title('Current file data')
ym = 2e-12
ylim(-ym,-0.5*ym)
```

### Working with data base

```python
coursor = gps_db.find()
gps_data = []
for item in coursor:
#     print(item.keys())
    gps_data.append(item)
print(len(gps_data))
```

```python
fs = []
fs_err = []
for dd in gps_data:
#     print(dd['data'])
    d = array([tuple(x) for x in dd['data']],dtype=[('n','>i4'),('t_cpu','>i4'),('dt','>i4'),('df','>f4')])
    print('df=%.1e err=%.1e for ' % (d['df'][-1],1 / (dd['freq']*1e3 * d['t_cpu'][-1])), str(dd['freq']/1e3) + '  ' + dd['t_start'])
    plot(d['n'],d['df'],label=str(dd['freq']/1e3) + '  ' + dd['t_start'])
    fs.append(d['df'][-1])
    fs_err.append(1 / (dd['freq']*1e3 * d['t_cpu'][-1])), str(dd['freq']/1e3)
legend(bbox_to_anchor=(1.8, 1))
xlabel('N shots (x100s)')
ylabel('Rel freq deviation')
title('Relative frequency stability')
ym = 2e-12
ylim(-ym,ym)
```
